```python
df_clean = df.drop(columns=['filename'])
df_clean = df_clean.apply(pd.to_numeric, errors='coerce')  
```

```python
df_clean = df_clean.dropna()  
```

```python
import lingam

model = lingam.DirectLiNGAM()
model.fit(df_clean)
print("Causal Order:", model.causal_order_)
```

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph(model.adjacency_matrix_)


plt.figure(figsize=(10, 6))
nx.draw_networkx(G, with_labels=True, node_color='skyblue', node_size=2000, font_size=12)
plt.title("Causal Graph from LiNGAM")
plt.show()
```

```python
import numpy as np
from castle.algorithms import ANMNonlinear

anm = ANMNonlinear()
anm.learn(df_clean)

causal_matrix = anm.causal_matrix
```

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph(causal_matrix)

plt.figure(figsize=(8, 6))
nx.draw_networkx(G, with_labels=True, node_color='lightgreen', node_size=2000, font_size=12)
plt.title("Neural Network-Based Causal Graph")
plt.show()
```
